## 2.시각화

### 1) 산점도 : plt.scatter, sns.scatterplot, sns.jointplot

① iris의 Petal.Length와 Petal.Width의 관계를 살펴보기 위해 산점도를 그려봅시다.
* plt.scatter

```python
plt.scatter(iris['Petal.Length'], iris['Petal.Width'])
plt.show()
```

* sns.scatterplot

```python
sns.scatterplot(data=iris, x='Petal.Length', y='Petal.Width')
plt.show()
```

* sns.scatterplot(    , hue = 'Species')

```python
sns.scatterplot(data=iris, x='Petal.Length', y='Petal.Width', hue='Species')
plt.show()
```

* sns.joinplot

```python
sns.jointplot(data=iris, x='Petal.Length', y='Petal.Width', hue='Species')
plt.show()
```

* 그래프로 부터 파악된 내용을 적어 봅시다.

> - length와 width 는 상관 관계가 있다.  
> - 길이와 넓이로 종을 구별할 수 있다.

### 2) 한꺼번에 산점도 그리기 : sns.pairplot

① air 데이터프레임에서 Month와 Day를 제외하고 산점도를 한꺼번에 그려봅시다.

```python
sns.pairplot(air, 
             x_vars=['Ozone','Solar.R','Wind','Temp'],
             y_vars=['Ozone','Solar.R','Wind','Temp'])
plt.show()
```

② iris 데이터프레임에 대해서, Species를 제외하고 한꺼번에 산점도를 그려봅시다.

```python
iris2 = iris.loc[:,['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']]
sns.pairplot(iris2)
plt.show()
```

## 3.수치화 : 상관분석

```python
import scipy.stats as spst
```

① boston.indus와 boston.medv의 관계를 수치화 해 봅시다.  

결과를 해석해 봅시다.
> 어느 정도 음의 상관관계를 가진다.  
> p값은 0에 가까운 값으로 귀무가설을 기각한다.


```python
spst.pearsonr(boston['indus'],boston['medv'])
```

② diamond의 각 변수들 간에 상관 계수를 구해 봅시다.

```python
d_cor = diamond.corr()
d_cor
```

```python
sns.heatmap(d_cor, 
            annot=True, fmt='.4f',
            vmax=1,vmin=-1)
plt.show()
```

③ 위 결과에서 가장 강한 상관관계와 약한 상관관계를 찾아 봅시다.

> 강한 상관관계 : x & y  
> 약한 상관관계 : depth & price
