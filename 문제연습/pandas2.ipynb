#### 상위 10개 행을 조회
 - "https://bit.ly/3FsgwkJ"

```python
titanic = pd.read_csv("https://bit.ly/3FsgwkJ")
titanic.head(10)
```

#### df1, df2 를 병합하시오.
 - df1 = pd.DataFrame({'col1':[1,2], 'col2':['a', 'b']})
 - df2 = pd.DataFrame({'col1':[3,4], 'col2':['c', 'd']})

```python
df1 = pd.DataFrame({'col1':[1,2], 'col2':['a','b']})
df2 = pd.DataFrame({'col1':[3,4], 'col2':['c','d']})

pd.concat([df1, df2], axis=1)
```

#### 숫자를 범주로 변환
 - 10 > Fare => L
 - 10 ~ 50 => M
 - 50 < Fare => H

```python
a = [-np.inf,10,50,np.inf]
b = ['L', 'M', 'H']

titanic['Fare_category'] = pd.cut(titanic['Fare'], bins=a, labels=b)
titanic
```
