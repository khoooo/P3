#### 1~100까지 숫자 중 홀수이면서 3의 배수를 리스트로 저장하시오.

```python
a = []
for i in range(1,101,1):
    if (i%2 ==1) & (i%3 == 0):
        a.append(i)
print(a)
```

```python
x1 = [i for i in range(1,101,1) if i%3 == 0 and i%2 == 1]
print(x1)
```

#### 리스트 이용하여 2차원 배열 생성

       - 전체
       - 열 기준

```python
a = [[1,2,3,4,5],[4,5,6,7,8]]

print(np.sum(a))
print(np.sum(a, axis=0))
print(np.sum(a, axis=1))
print('\n')
print(np.mean(a))
print('\n')
print(np.std(a, axis=1))
```

#### 데이터프레임을 불러와서 상위 10개 행 조회
 - 'https://bit.ly/3FsgwkJ'

```python
import pandas as pd

titanic = pd.read_csv('https://bit.ly/3FsgwkJ')
titanic.head(10)
```

#### 데이터프레임 1~5 행, Name, Age, Embarked 조회

```python
titanic[['Name','Age','Embarked']].head(5)
```

#### Age가 70세 이상인 탑승객 조회

```python
titanic[titanic['Age'] >= 70]
```

#### Age가 60세 이상이고, 여성인 탑승객 조회

```python
a = titanic['Age'] >= 60
b = titanic['Sex'] == 'female'
titanic[a & b]
```

#### Pclass가 1이거나 3인 탑승객 조회

```python
titanic[titanic['Pclass'].isin([1,3])]
```

#### 승선지역(Embarked)별 평균 운임을 조회

```python
titanic.groupby('Embarked')['Fare'].mean()
```
