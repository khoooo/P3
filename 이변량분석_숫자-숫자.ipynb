```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as spst
import statsmodels.api as sm

from statsmodels.graphics.mosaicplot import mosaic
```

```python
boston=pd.read_csv('https://bit.ly/3EuWvZw')
boston.head(3)
```

## 숫자형 - 숫자형

> - 시각화 : 산점도 

> - 검증 : 상관관계 분석

#### crim(범죄율) - medv(집값)

```python
sns.scatterplot(data=boston, x='crim', y='medv')

plt.show()
```

```python
sns.jointplot(data=boston, x='crim', y='medv')
plt.show()
```

#### 전체 산점도

```python
sns.pairplot(boston)
plt.show()
```

#### 개수 골라서 할 수도 있음.

```python
boston_loc = boston.loc[:,['crim','indus','nox','medv']]

sns.pairplot(boston_loc)
plt.show()
```

#### 검증

```python
spst.pearsonr(boston['crim'], boston['medv'])
```

```python
boston.corr()
```

```python
plt.figure(figsize=(16,10))

sns.heatmap(boston.corr(), annot=True,
           vmax=1, vmin=-1,
           fmt='.4f')
plt.show()
```

#### 상관분석 검증과정

- 0.5 이상 : 강한 양의 상관관계


- -0.5 이하 : 강한 음의 상관관계


- -0.5< < 0.5 : 그래프와 같이 비교
