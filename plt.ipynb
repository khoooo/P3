```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### 다음 데이터를 사용하여 라인차트 그리시오.
 - 데이터 : [2,5,3,1,2]

```python
a = [2,5,3,1,2]
plt.plot(a)
plt.show()
```

#### x,y 지정하고 라인차트 그리시오.
 - x : 1,2,3,4,5
 - y : 2,5,3,1,2

```python
x = [1,2,3,4,5]
y = [2,5,3,1,2]
plt.plot(x,y)
plt.show()
```

#### 딕셔너리 만들고 라인차트 그리시오.
 - v1:[1,2,3,4,5]
 - v2:[2,5,3,1,2]

```python
dict1 = {'v1':[1,2,3,4,5], 'v2':[2,5,3,1,2]}

plt.plot('v1','v2',data=dict1)
plt.show()
```

#### 조건에 맞게 그리시오.
 - x 축 레이블 : month
 - y 축 레이블 : sales
 - 타이틀 : Monthly Sales
 - color : green
 - marker : circle
 - line style : dashed

```python
plt.plot('v1','v2','go--', data=dict1)

plt.xlabel('month')
plt.ylabel('sales')
plt.title('Monthly Sales')

plt.show()
```

#### 위 조건과 동일하게 2개의 그래프를 그리시오.
 - v1 : ['1,2,3,4,5]
 - v2 : [2,5,3,1,2]
 - v3 : [4,6,7,6,5]

```python
dict2 = {'v1':[1,2,3,4,5], 'v2':[2,5,3,1,2], 'v3':[4,6,7,6,5]}

plt.plot('v1','v2','go--', data=dict2)
plt.plot('v1','v3','ys-', data=dict2)

plt.xlabel('month')
plt.ylabel('sales')
plt.title('Monthly Sales')

plt.show()
```

#### 위의 조건에서 grid, legend 를 추가하시오.
 - v2 : apple
 - v3 : mongo

```python
plt.plot('v1','v2','go--',data=dict2,label='apple')
plt.plot('v1','v3','rv-',data=dict2,label='mongo')

plt.xlabel('month')
plt.ylabel('sales')
plt.title('Monthly sales')

plt.grid()
plt.legend()

plt.show()
```

#### 축 범위를 추가하시오.
 - x : 0~6
 - y : 0~8

```python
plt.plot('v1','v2','go--',data=dict2,label='apple')
plt.plot('v1','v3','rv-',data=dict2,label='mongo')

plt.xlabel('month')
plt.ylabel('sales')
plt.title('Monthly sales')

plt.grid()
plt.legend()

plt.xlim(0,6)
plt.ylim(0,8)

plt.show()
```

#### 수평선과 수직선을 추가하시오.
 - 수평선 : 5.3
 - 수직선 : 2.4

```python
plt.plot('v1','v2','go--',data=dict2,label='apple')
plt.plot('v1','v3','rv-',data=dict2,label='mongo')

plt.xlabel('month')
plt.ylabel('sales')
plt.title('Monthly sales')

plt.grid()
plt.legend()

plt.xlim(0,6)
plt.ylim(0,8)

plt.axhline(5.3, color='gray', linestyle='--')
plt.axvline(2.4, color='gray', linestyle='--')
plt.text(1,5.5,'5.3')
plt.text(2.1,7,'2.3')

plt.show()
```

#### 그래프를 나눠서 그리시오.
 - 1x2

```python
plt.subplot(1,2,1)
plt.plot('v1','v2',data=dict2)

plt.subplot(1,2,2)
plt.plot('v1','v3',data=dict2)

plt.show()
```

#### 그래프를 나눠서 그리시오.
 - 2x2

```python
plt.subplot(2,2,1)
plt.plot('v1','v2','go--',data=dict2)

plt.subplot(2,2,2)
plt.plot('v1','v3',data=dict2)

plt.subplot(2,2,3)
plt.plot('v1','v3',data=dict2)

plt.subplot(2,2,4)
plt.plot('v1','v2','y*-',data=dict2)

plt.show()
```
